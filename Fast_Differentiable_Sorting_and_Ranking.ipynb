# Fast Differentiable Sorting and Ranking

This colab notebook treats the following paper:

> [Blondel, Mathieu, et al. "Fast Differentiable Sorting and Ranking." arXiv  preprint arXiv:2002.08871 (2020)](https://arxiv.org/pdf/2002.08871).

Their paper proposes the first *differentiable sorting and ranking operators* with $\BigO(n\log(n))$ time and $\BigO(n)$ memory complexity. Furthermore, their approach enjyos *exact* computation and differentiation by casting differentiable sorting and ranking into projections onto the permutahedron (the convex hull of all permutations) and using a reduction to isotonic optimization (linear programming).

For their implementation have a look at:

> https://github.com/google-research/fast-soft-sort

# Math

**Def. (Scores/Logits)** Let $\vtheta:=(\theta_1,\ldots,\theta_n)$ denote a vector of scores or "logits" produced by a model, i.e.,
$$
\vtheta:=g(\vx),
$$
for some model $g\colon\cX\to\R^n$ and some $\vx\in\cX$.

**Def. (Set of Numbers from 1 to $n$)** $[n]:=\set{1,\ldots,n}$

**Note.** The set does not enforce any order!

**Def. (Permutation of $[n]$)** A permutation of $[n]$ is denoted by $\sigma=(\sigma_1,\ldots,\sigma_n)$ and its inverse is denoted by $\sigma^{-1}$.

**Note.** We can describe $\sigma$ as *ordered* tuple that contains all elements of $[n]$ exactly once.

**Com.** For convenience we will sometimes use the following conventions:
- Inverse permutation $\pi:=\sigma^{-1}$
- If a permutation $\sigma$ is interpreted as a vector, we denote it with bold, $\vsigma\in[n]^n$
- One may also interpret $\sigma$ as a bijection over $[n]$.
- The permutation $(1,2,\ldots,n)$ is called the *identity* permutation.
- The *reversing* permutation is denoted as $\rho=(n,n-1,\ldots,1)$.

**Def. (Set of Permutations of $[n]$)** The set of the $n!$ permutations of $[n]$ is denoted as $\Sigma$.

**Def. (Permutation of a Vector)** Given a permutation $\sigma\in\Sigma$, we denote the version of a vector $\theta=(\theta_1,\ldots,\theta_n)\in\R^n$ permuted according to $\theta$ by
$$
\vtheta_{\sigma}
:=
(\theta_{\sigma_1},\ldots,\theta_{\sigma_n})\in\R^n.
$$

**Def. (Restriction of a Vector to a Set of Indices)** Given a set $\cS\subseteq[n]$ and a vector $\vv\in\R^n$, we denote the *restriction of $\vv$ to $\cS$* by
$$
\vv_{\cS}:=(v_i\colon i\in S)\in\R^{\card{S}}
$$
TODO: is there any order here???

**Def. (Argsort)** A *argsort $\sigma(\vtheta)$* of a vector $\vtheta\in\R^n$ is a permutation $\sigma\in[n]^n$ 
$$
\sigma(\vtheta)
:=
(\sigma_1(\vtheta),\ldots,\sigma_n(\vtheta))\in[n]^n
$$
that sorts the coefficients of $\vtheta$ in descending order:
$$
\theta_{\sigma_1(\vtheta)}\geq \theta_{\sigma_2(\vtheta)}\geq \cdots\geq \theta_{\sigma_n(\vtheta)},
$$
where ties are broken arbitrarily.

**Def. (Sort)** A *sort* of a vector $\vtheta\in\R^n$ are the coefficients of $\vtheta$ in descending order, i.e.,
$$
s(\vtheta)=\vtheta_{\sigma(\vtheta)}
$$
where $\theta(\vtheta)$ is an argsort of $\vtheta$.

**Def. (Rank)** A rank of a vector $\vtheta\in\R^n$ is the inverse permutation of an argsort of $\vtheta$:
$$
r(\vtheta)=\sigma^{-1}(\vtheta).
$$
A rank of $\vtheta$ is a "function" $[n]\to[n]$ that evaluates at each coordinate $j$ (from $1$ to $n$) to the position $k\in[n]$ that $\theta_j$ lands in the corresponding descending sort of the argsort. That means that the smaller the rank $r_j(\vtheta)$ the higher the value of $\theta_j$ is.



We use the descending order for argsort, sort and rank for convenience. The ascending counterparts could be easily obtained through
$$
\sigma(-\vtheta),\quad-s(-\vtheta),\quad r(-\vtheta),
$$
respectively.

**Example**

$$
\vtheta=(\theta_1,\theta_2,\theta_3)\quad\text{with }\theta_3\geq\theta_1\geq\theta_2.
$$

An argsort $\sigma(\vtheta)\in[n]^n$ of $\theta$ is: 
$$
\sigma(\vtheta)=(3,1,2).
$$
The corresponding sort to the argsort is: 
$$
s(\vtheta)=\vtheta_{\sigma(\vtheta)}=(\theta_3,\theta_1,\theta_2).
$$
The corresponding rank of the argsort is:
$$
r(\vtheta)=(2,3,1).
$$

## Sorting and Ranking as Linear Programs

Here we show how to cast sorting and ranking operations as linear programs over the permutahedron. To that end, we first formulate the argsort and ranking operations as *discrete* optimization problems over the set of permutations $\Sigma$ (we'll reformulate these as *continuous* optimization problems later).

**Lem.** For all $\vtheta\in\R^n$ and $\rho=(n,n-1,\ldots,1)$ we have
$$
\begin{align*}
\sigma(\vtheta)&\in\argmax_{\sigma\in\Sigma}\scprod{\vtheta_{\sigma},\rho},\quad\text{and}
\\
r(\vtheta)&\in\argmax_{\pi\in\Sigma}\scprod{\vtheta,\rho_{\pi}}.
\end{align*}
$$








TODO: say why this is a linear program!

**Proof.** In other words, to get the argsort $\sigma(\vtheta)$, we seek a permutation $\sigma$ that maximizes the inner product $\scprod{\vtheta_{\sigma},\rho}$, which can also be written as
$$
\sigma(\vtheta)\in\argmax_{\sigma\in\Sigma}
n\cdot\theta_{\sigma_n(\vtheta)}
+
(n-1)\cdot\theta_{\sigma_{n-1}(\vtheta)}
+\cdots+
1\cdot\theta_{\sigma_1(\vtheta)}.
$$
This clearly shows that it is desirable that $\sigma$ moves the highest value of $\vtheta$ to the front of the sum, then the second largest, etc. Thus $\sigma$ sorts $\vtheta$ from highest to lowest.

Further, be aware of the rule that for any two vectors $\vx,\vy\in\R^n$ and any permutation $\sigma\in\Sigma$, we have
$$
\scprod{\vx_\sigma,\vy}=\scprod{\vx,\vy_{\sigma^{-1}}}.
$$
With this rule, and the definition that $r(\vtheta)=\sigma^{-1}(\vtheta)$, we easily get to:
$$
\begin{align*}
r(\vtheta)&\in\argmax_{\pi\in\Sigma}\scprod{\vtheta_{\pi^{-1}},\rho}
\\
&=\argmax_{\pi\in\Sigma}\scprod{\vtheta,\rho_{\pi}}.
\end{align*}
$$

**Def. (Permutahedron induced by a Vector)** To obtain a *continuous* optimization problem, we introduce the permutahedron, induced by a vector $\vw\in\R^n$, which is the *convex hull* of the set of permutations $\dset{\vw_{\sigma}}{\sigma\in\Sigma}$ of $\vw$:
$$
\cP(\vw)=
\text{conv}\left(\dset{\vw_{\sigma}}{\sigma\in\Sigma}\right)
=
\dset{
\sum_{i=1}^{n!}c_i\vw_{\sigma^{(i)}}
}{
\sum_{i=1}^{n!}c_i=1 \,\land\, \left(\bigwedge_{i=1}^{n!} c_i\geq 0\right)
}\subset\R^n,
$$
where we number the permutations from $\sigma^{(1)}$ to $\sigma^{(n!)}$ (as opposed to referring to their coordinates by $\sigma_i$).

Any permutahedron induced by a vector $\vw$ is a *convex polytope*, whose vertices correspond to permutations of $\vw$. In particular, when $\vw=\rho$ (or $\vw=(1,2,\ldots,n)$) we have that $\cP(\vw)=\text{conv}(\MSigma)$.

The permutahedron $\cP((1,2,3))\subset\R^3$ is illustrated in the figure below:

![Permutahedron induced by $\vw=(1,2,3)$](https://upload.wikimedia.org/wikipedia/commons/thumb/6/6e/Permutohedron_order_3.svg/600px-Permutohedron_order_3.svg.png)

Having defined the upper optimization problem and the permutahedron we can redefine the optimzation problems of above as *continuous* optimization problems, that give linear programming formulations for sort and rank.

**Prop.** For all $\vtheta\in\R^n$ and $\rho=(n,n-1,\ldots,1)$ we have
$$
\begin{align*}
s(\vtheta) &\in \argmax_{\vv\in\cP(\vtheta)}\scprod{\vv,\rho}, \text{ and}
\\
r(\vtheta) &\in \argmax_{\vv\in\cP(\rho)}\scprod{-\vtheta,\vv}.
\end{align*}
$$

This just follows from the fact that $\theta_\sigma\in\cP(\vtheta)$ (special case: vertice of convex polytope) and the fact that an optimal solution of a linear program is always achieved at a vertex of the convex polytope. Thus, the optimal solution over a permutahedron is always a permutation.

The interesting thing to note in the upper proposition is that for $s(\vtheta)$
- $\vtheta$ (the function "logits) *appears in the constraints*, and 
- $\vrho$ *appears in the objective*.

While for $r(\vtheta)$ it is the other way around:
- $\vrho$ *appears in the constraints*, and 
- $\vtheta$ (the function "logits) *appears in the objective*.



**Proof** (for ranking, since sorting easily follows from absorbing the permutation in the formula for $\sigma(\vtheta)$)
$$
\begin{align*}
r(\vtheta)
&\in\argmax_{\pi\in\Sigma}\scprod{\vtheta,\rho_{\pi}}
\\
&=\argmax_{\pi\in\Sigma}\scprod{\vtheta_{\pi^{-1}},\rho}
\\
&=\argmin_{\pi\in\Sigma}\scprod{\vtheta_{\pi^{-1}},(1,2,\ldots,n)}
\\
&=\argmax_{\pi\in\Sigma}\scprod{-\vtheta_{\pi^{-1}},(1,2,\ldots,n)}
\\
&=\argmax_{\pi\in\Sigma}\scprod{-\vtheta,(1,2,\ldots,n)_{\pi}}
\\
&=\argmax_{\pi\in\Sigma}\scprod{-\vtheta,\pi}
\\
&=\argmax_{\vv\in\cP(\vrho)}\scprod{-\vtheta,\vv}.
\end{align*}
$$

Another nice thing to note is that the upper proposition contains linear programs:

**Def. (Linear Program)** A *linear program* is an optimization problem that

- has a *convex polytope* as a solution space (constraints)
- an an objective function that is *linear* in terms of the parameter that is optimized in the solution space.

$$
\begin{matrix}
\\
\\
\boxed{\text{opt. sol.}}\\
\text{vertice of}\\
\text{conv. polytope}
\end{matrix}
=
\argmax_{
\begin{matrix}
\boxed{constraint}\\
\text{conv. polytope}
\end{matrix}
}
\begin{matrix}
\\
\\
\boxed{\text{objective func.}}\\
\text{linear in terms of param.}\\
\text{on conv. polytope}
\end{matrix}
$$

Furthermore, the solution will always lie on a vertice of the convex polytope.

# Permutahedron Example

Let's have a look how the maximization objective looks like on various 2D and 3D Permutahedrons.

## Plotting Function

```
def create_permutahedron_plot(theta=[0.5, 0.2, 1.2], eps=0.0, resolution=20):

    dim = len(theta)
    
    # create theta
    if dim == 2:
        theta = np.array(theta)
    elif dim == 3:
        theta = np.array(theta)
    else:
        raise ValueError('Only 2 or 3 dimensions are supported!')
    
    # generate permutations
    P = np.array(list(itertools.permutations(range(dim)))).T

    # create permuted versions of theta
    T = np.zeros(P.shape)
    for i in range(P.shape[1]):
        T[:, i] = theta[P[:, i]]

    def _gen_coefficients(current=[], num_coefficients=-1):
        length = len(current)
        sum_so_far = sum(current)
        if length == num_coefficients - 1:
            c = 1.0 - sum_so_far
            current.append(c)
            yield current.copy()
            current.pop()
        elif length < num_coefficients - 1:
            for i in range(resolution + 1):
                c = (resolution - i) / float(resolution)
                if sum_so_far + c <= 1.0:
                    current.append(c)
                    yield from _gen_coefficients(current, num_coefficients)
                    current.pop()

    def gen_coefficients(num_coefficients):
        L = list(_gen_coefficients(current=[], 
                                   num_coefficients=num_coefficients))
        C = np.array(L).T
        return C

    # generate convex combination coefficients
    C = gen_coefficients(math.factorial(dim))
    C.shape
    # compute convex combinations of theta's permutations
    V = np.matmul(T, C)
    # generate reverse permutation vector
    r = np.expand_dims(np.array(range(dim, 0, -1)), axis=1)
    R = np.repeat(r, V.shape[1], axis=1)

    # compute gain (value of maximization obj. of linear program) 
    # = of inner prod. with reverse perm.
    g_sort = (V * R).sum(axis=0)

    # compute regularization term (if eps > 0)
    if eps > 0.0:
        Q_V = 0.5 * (V * V).sum(axis=0)
        g_sort -= eps * Q_V

    # find index of convex combination of permutations with maximal gain
    i_max = np.argmax(g_sort)

    # create plot
    fig = plt.figure(figsize=(16, 12))
    # determine gain range for colormap
    g_min = g_sort.min()
    g_max = g_sort.max()
    g_rel = (g_sort - g_min) / (g_max - g_min)
    # create color map
    colmap = cm.ScalarMappable(cmap=cm.plasma)
    colmap.set_array(g_rel)
    if dim == 2:
        ax = fig.gca()
        xs, ys = V[0, :], V[1, :]
        yg = ax.scatter(xs, ys, c=cm.plasma(g_rel), marker='o', s=100, alpha=0.5, zorder=10)
        yg_max = ax.scatter(xs[i_max], ys[i_max], 
                            c='#000000', marker='*', s=200, zorder=50)
        
    elif dim == 3:
        ax = fig.add_subplot(111, projection='3d')
        xs, ys, zs = V[0, :], V[1, :], V[2, :]
        yg = ax.scatter(xs, ys, zs, c=cm.plasma(g_rel), marker='o', s=100, alpha=0.2, zorder=10)
        yg_max = ax.scatter(xs[i_max], ys[i_max], zs[i_max], 
                            c='#000000', marker='*', s=200, zorder=50)
        ax.view_init(20, 30)
    # add permutation labels
    for i in range(P.shape[1]):
        p = P[:, i]
        pp = T[:, i]
        txt = str(pp.tolist()) + '\n' + 'σ=' + str((1+p).tolist()) 
        ax.text(*pp, txt, fontsize=10, ha='center', weight='bold', zorder=100)

    # TODO: add optimal convex comb. permutation label and values

    cb = fig.colorbar(colmap)
    title = 'Heatmap of Linear Program Sorting Maximization Objective\n'\
            'on Permutahedron Induced by θ='+str(theta.tolist())+'\n'\
            'ε='+str(eps)
    plt.title(title)
    plt.show()
```

```
def test_pearsonr():
    for i in range(1000):
        logits = [random.random() for _ in range(100)]
        targets = [i/100 for i in range(100)]
        logits = torch.tensor(logits, dtype=torch.float64)
        targets = torch.tensor(targets, dtype=torch.float64)
        np_corr = np.corrcoef(logits, targets)[0, 1]
        pt_corr = pearsonr(logits, targets)[0].item()
        assert abs(np_corr - pt_corr) < 1e-10

# numerical check that np_corr is equivalent to pearsonr
test_pearsonr()
```

```
def numerai_score(logits_list, targets_list):
    df_logits = pd.DataFrame(logits_list)
    df_targets = pd.DataFrame(targets_list)
    ranks_pct = np.ravel(df_logits.rank(pct=True, method='first').values)
    targets = np.ravel(df_targets.values)
    corr = np.corrcoef(ranks_pct, targets)[0, 1]
    return ranks_pct, corr

def pytorch_score(logits_list, targets_list):
    logits = torch.tensor(logits_list, dtype=torch.float64)
    targets = torch.tensor(targets_list, dtype=torch.float64)
    logits_extra = logits.unsqueeze(0)
    ranks_ext = soft_rank(logits_extra,
                          direction='ASCENDING',
                          regularization_strength=0.05,
                          regularization='l2')
    ranks = ranks_ext.squeeze(0)
    # convert ranks to percentage scale
    # (the same way that pandas does it)
    ranks_pct = ranks / ranks.shape[0]
    corr = pearsonr(ranks_pct, targets)
    return ranks_pct, corr

def printres(logits_list, l, c):
    print('[', end='')
    for v in l:
        print(f'{v:.2f}, ', end='')
    print(f'] {c:.2f}')

# if there are no ties in the logits (which tere likely wont be), 
# then the pytorch score behaves exactly as the numerai score
logits_list = [0.3, 0.15, 0.5, 0.35, 0.1]
targets_list = [0.1, 0.2, 0.3, 0.1, 0.1]
nmr_ranks_pct, nmr_corr = numerai_score(logits_list, targets_list)
pt_ranks_pct, pt_corr = pytorch_score(logits_list, targets_list)
pt_ranks_pct = pt_ranks_pct.tolist()
pt_corr = pt_corr[0].item()
print(logits_list)
printres(logits_list, nmr_ranks_pct, nmr_corr)
printres(logits_list, pt_ranks_pct, pt_corr)
```
